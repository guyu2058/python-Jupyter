```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
 
```

# Pandas的汇总 
- 级联： pd.concat, pd.append    
- 合并： pd.merge, pd.join
  **索引使用字符做列表，要使用list（）**

以索引对齐的方式实现

```python
score1= DataFrame(np.random.randint(0,100,size=(3,3)), columns=["python","java","php"],index=list("abc"))
score2= DataFrame(np.random.randint(0,100,size=(5,3)), columns=["python","java","php"],index=list("deabc"))
display(score1,score2)
```

```python
#objs 表示要参与级联的df的对象的列表，或者元祖
pd.concat(objs=(score1,score2))
```

```python
# 级联后， 当索引值在原始表中没有意义的时候， 可以忽略，避免重复
pd.concat(objs=(score1,score2),ignore_index=True)
```

```python
#axis 改变级联的方向
pd.concat(objs=(score1,score2),axis=1)
```

```python
#axis 改变级联的方向后，
# 使用keys 可以增加多层列标签
pd.concat(objs=(score1,score2),axis=1,keys=["第一学期","第二学期"])
```

- 对相同结构的数据进行汇总， 或者比较不同时期的数据，可以使用级联操作
- 对不同结构的数据，通过关联项（关键项）进行合并，形成整体大表
- join包含（内连接与外连接）

```python
score1, score2
```

```python
# 增加一列C语言的科目
CC =pd.DataFrame(data=np.random.randint(0,100,size=(3,1)),index=list("abc"),columns=["C"])
CC
score3=pd.concat(objs=(score1,CC),axis=1)
score3
score4=pd.concat(objs=(score2,CC),axis=1)
score4
```

```python
#增加一名D同学
dd = DataFrame(data=np.random.randint(0,100,size=(1,4)),index=["d"],columns=score3.columns)
dd
score5 = pd.concat(objs=(score3,dd))
score5
```

```python
display(score1,score5)
```

```python
# outer连接 或运算
pd.concat(objs=(score4,score5),join="outer",axis=1)
```

```python
# inner连接  并运算
pd.concat(objs=(score4,score5),join="inner",axis=1)
```

```python
#左连接，还是右连接，以df1,还是df2
# "join_axes"在0.25版中已弃用
pd.concat(objs=(score1,score5),join_axes=[score1.columns])
```

#  pd.merge()的合并 
- merge与concat的区别，合并只针对列操作，依据共同的列进行合并
- 参与合并的两个表，存在至少一列，或者多列， 有对应关系（一对一、一对多，多对多）  
- 
- 

```python
table1= pd.read_excel("案例.xlsx",sheet_name=1)
table2= pd.read_excel("案例.xlsx",sheet_name=2)
table3= pd.read_excel("案例.xlsx",sheet_name=3)
table4= pd.read_excel("案例.xlsx",sheet_name=4)
```

```python
display(table1 ,table2,table3,table4)
```

```python
# 一对一 合并 ,默认默认找相同项， 这里默认为序号，
pd.merge(table1, table2)
```

```python
#多对一合并,制定针对的列进行合并，通过选择“‘’姓名”
pd.merge(table1,table3,on="姓名")
```

```python
#多对多关系
pd.merge(table3,table4,on=["学科","姓名"])
# 
pd.merge(table3,table4,on=["姓名"],suffixes=["_上学期","_下学期"])
```

```python
#分别指定合并的列标签 
pd.merge(table2,table3,left_on="学科二",right_on="学科")
# 不显示多余的列项
pd.merge(table2,table3,left_on="学科二",right_on="学科").drop(labels=["学科二","学科三"], axis=1)
```

```python
#how默认是ininer内连接， 选择outer是外连接，
#还有left 和right 表示保留左表还是右表的内容
#内连接是是显示相同的内容， 过滤掉不同的内容

pd.merge(table1,table2,how="outer")
```
