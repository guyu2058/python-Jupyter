# 1 级联，两个数据连接
- 

```python
arr1 = np.random.randint(0,10,size=(3,3))
arr2 = np.random.randint(11,20,size=(3,3))
display(arr1,arr2)
```

```python
#默认纵向链接，参与的列数要一致
np.concatenate((arr1,arr2))
```

```python
#横向纵向链接 ,参与数组的行数要一致
np.concatenate((arr1,arr2),axis=1)
```

## 2 切分 ：
- np.split  
- np.vsplit  
- np.hsplit  

```python
arr = np.random.randint(0,100,size=(6,6))
arr
```

```python
#指定切割的份数
arr1,arr2 = np.split(arr,indices_or_sections=2)
display (arr1,arr2)
```

```python
#使用axis调整切割方向
arr3,arr4 = np.split(arr,indices_or_sections=2,axis=1)
display (arr3,arr4)
```

```python
# indices_or_sections=[m,n]切换数量
arr3,arr4,arr5 = np.split(arr,indices_or_sections=[2,5],axis=1)
display (arr3,arr4,arr5)
```

```python
arr3,arr4,arr5 = np.hsplit(arr,indices_or_sections=[2,5])
display (arr3,arr4,arr5)
```
